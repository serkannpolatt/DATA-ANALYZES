Task #1: What percentage of all licensed sets ever released were Star Wars Themed?

-Görev 1: Şimdiye kadar piyasaya sürülen tüm lisanslı setlerin yüzde kaçı Star Wars Temalıydı?-

```python
merged=data.merge(theme, left_on="parent_theme",right_on="name")
```

```python
merged.head()
```

```python
merged.drop(columns="name_y",inplace=True)
```

```python
merged.head()
```

```python
licensed=merged[merged["is_licensed"]]
```

```python
licensed.head()
```

```python
licensed.plot()
```

```python
star_wars=licensed[licensed["parent_theme"]=="Star Wars"]
```

```python
star_wars.head(50)
```

```python
star_wars.plot()
```

```python
star_wars.plot.hist(alpha=0.7)
```

```python
the_force=int(star_wars.shape[0]/licensed.shape[0]*100)
```

```python
print(the_force)
```

```python
merged[merged["set_num"].isnull()].shape
```

```python
licensed=merged[merged["is_licensed"]]
licensed=licensed.dropna(subset=["set_num"])

star_wars=licensed[licensed["parent_theme"]=="Star Wars"]
the_force=int(star_wars.shape[0]/licensed.shape[0]*100)
```

```python
print(the_force)
```

Task #2: In which year was Star Wars not the most popular licensed theme?

-Görev #2: Star Wars hangi yılda en popüler lisanslı tema değildi?-

```python
licensed_sorted=licensed.sort_values("year")
```

```python
licensed_sorted
```

```python
licensed_sorted.groupby(["year","parent_theme"]).head(50)
```

```python
licensed_sorted["count"]=1
licensed_sorted.head()
```

```python
licensed_sorted.groupby(["year","parent_theme"]).sum()
```

```python
summed_data=licensed_sorted.groupby(["year","parent_theme"]).sum().reset_index()
```

```python
summed_data.head()
```

```python
max_data=summed_data.sort_values("count",ascending=False).drop_duplicates(["year"])
```

```python
max_data.head()
```

```python
max_data.sort_values("year",inplace=True)
```

```python
max_data.plot.hist(alpha=0.7)
```
