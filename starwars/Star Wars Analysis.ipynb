```python
data=pd.read_csv("movie_metadata_modificado3.csv")
```

To get a sense of what our data file looks like, let us first take a peak at the first 5 rows. Then we will print out information descrbing all of the columns of the data set.

-Veri dosyamızın neye benzediğini anlamak için önce ilk 5 satırda bir tepe yapalım. Ardından, veri setinin tüm sütunlarını açıklayan bilgileri yazdıracağız.

```python
# Find the values contained in the "Star Wars" group
#"Yıldız Savaşları" grubunda yer alan değerleri bulun

df=pd.DataFrame(data,columns=["movie_title","director_name","title_year","budget","gross","imdb_score"])
star_wars=df[data["movie_title"].str.contains("Star Wars")]
star_wars_sorted=star_wars.sort_values(by=["title_year"])
star_wars_sorted
```

Ok great. We have most of what we need, except that Episode VII - The Force Awakens is missing values and Episode VIII - The Last Jedi is not in the data set. In addition, Star Wars - The Clone Wars is listed, which is a 3D animated movie that is not included in the trilogies. We will thus drop it from our list. There are two other Star Wars movies, "Rogue One", and "Solo", which we will not include in our analysis for the same reason that we are not including The Clone Wars.

-Tamam harika. Bölüm VII - Güç Uyanıyor'un eksik değerler olması ve Bölüm VIII - Son Jedi'ın veri setinde olmaması dışında ihtiyacımız olan şeylerin çoğuna sahibiz. Ayrıca üçlemelerde yer almayan bir 3D animasyon filmi olan Star Wars - The Clone Wars da listeleniyor. Bu nedenle listemizden çıkaracağız. Klon Savaşları'nı dahil etmememizle aynı nedenle analizimize dahil etmeyeceğimiz iki Star Wars filmi daha var, "Rogue One" ve "Solo".

```python
# Drop the titles with missing values
# Eksik değerlere sahip başlıkları bırakın

star_wars_drop=star_wars_sorted.dropna()
star_wars_drop
```

After dropping all of the rows that had missing values (Episode VII - The Force Awakens and The Clone Wars), I went to IMDB.com to get updated information for Episode VII and VIII. We will manually add in those movies as 2 new rows with their respective column attributes.

-Eksik değerleri olan tüm satırları (Bölüm VII - The Force Awakens ve The Clone Wars) bıraktıktan sonra, Bölüm VII ve VIII için güncel bilgileri almak için IMDB.com'a gittim. Bu filmleri, ilgili sütun nitelikleriyle birlikte 2 yeni satır olarak manuel olarak ekleyeceğiz.

```python
# Add 2 new rows: Episode VII and VIII, which were not adequately represented or present in our imdb dataset
# 2 yeni satır ekleyin: imdb veri setimizde yeterince temsil edilmeyen veya mevcut olmayan Bölüm VII ve VIII

VII_VIII=[pd.Series(["Star Wars: Episode VII - The Force Awakens?","JJ. Abrams",2015.0,245000000.0,936662225.0, 8.0],index=star_wars_drop.columns),
         pd.Series(["Star Wars: Episode VIII - The Last Jedi?","Rian Johnson",2017.0, 317000000.0, 620181382.0, 7.1], 
                      index=star_wars_drop.columns)]

star_wars_full=star_wars_drop.append(VII_VIII,ignore_index=True)
```

Now that we have our complete list of Star Wars movies with all of the necessary data, we need to reset the index to (0-7). Resetting the index makes plotting the movies in this specific order possible.

-Artık gerekli tüm verilerle birlikte tam Star Wars film listemize sahip olduğumuza göre, dizini (0-7) olarak sıfırlamamız gerekiyor. Dizinin sıfırlanması, filmlerin bu özel sırayla çizilmesini mümkün kılar.

```python
star_wars_index=star_wars_full.reset_index(drop=True)
star_wars_index
```

Looks perfect, except that all of the movies have a question mark (?) at the end of their title. It doesn't bother me so much to see it in the dataframe above, but when I tried to make my visualizations with the question marks present, it did not look so great. Thus, we will have to do some more df-manipulation to get it looking right. The best way I could find to do so was to recreate the movie_title column as (movietitle) and to rewrite the movie names without the question marks. We can then drop the original movie_title column and re-arrange the columns so that movietitle is now in front.

-Tüm filmlerin başlıklarının sonunda bir soru işareti (?) olması dışında mükemmel görünüyor. Yukarıdaki dataframe'de görmek beni çok rahatsız etmiyor, ancak görselleştirmelerimi soru işaretleriyle yapmaya çalıştığımda pek harika görünmüyordu. Bu nedenle, doğru görünmesini sağlamak için biraz daha df-manipülasyonu yapmamız gerekecek. Bunu yapmanın en iyi yolu movie_title sütununu (film başlığı) olarak yeniden oluşturmak ve film adlarını soru işaretleri olmadan yeniden yazmaktı. Daha sonra orijinal movie_title sütununu bırakabilir ve sütunları, movietitle şimdi önde olacak şekilde yeniden düzenleyebiliriz.

```python
#remove unwanted ? at the end of the movie titles by creating a new column 
#istenmeyen kaldırılsın mı? film başlıklarının sonunda yeni bir sütun oluşturarak

star_wars_clean=star_wars_index.assign(movie_title_=star_wars_index["movie_title"].str.replace(r"?",""))

#drop original movie_title column containing the movie titles with ?
#ile film başlıklarını içeren orijinal movie_title sütununu ?

star_wars_drop2=star_wars_clean.drop(["movie_title"],axis=1)
star_wars_drop2
```

```python
#re-arrange columns so that movie_title_ is first
#film_başlığı_ ilk olacak şekilde sütunları yeniden düzenleyin

cols=star_wars_drop2.columns.tolist()
cols=cols[-1:] + cols[:-1]
cols
```

```python
#rename cols to sw_arranged and display as a df
# sütunları sw_arranged olarak yeniden adlandırın ve bir df olarak görüntüleyin

star_wars_arragned=star_wars_drop2[cols]
star_wars_arragned
```

Ok perfect. We are just about ready to start our analysis. I decided to streamline the columns a little further so that only the columns we will use for analysis are present in the dataframe. It's not something that is absolutely neccessary, but for simplicity sake we will do so.

-Tamam mükemmel. Analizimize başlamak için hemen hemen hazırız. Veri çerçevesinde yalnızca analiz için kullanacağımız sütunların bulunması için sütunları biraz daha düzene sokmaya karar verdim. Bu kesinlikle gerekli bir şey değil, ancak basitlik adına bunu yapacağız.

```python
#remove director name and title year to leave only the columns used for analysis
#Yalnızca analiz için kullanılan sütunları bırakmak için yönetmen adını ve unvan yılını kaldır

star_wars_budget_gross=star_wars_arragned[["movie_title_",'budget','gross', 'imdb_score']]
star_wars_budget_gross_plot = star_wars_budget_gross.sort_index(ascending=False)
star_wars_budget_gross_plot
```

Time to start making some plots. Let's start by making two horizontal bar plots, one for budget and one for gross.

-Bazı planlar yapmaya başlama zamanı. Biri bütçe, diğeri brüt olmak üzere iki yatay çubuk grafiği oluşturarak başlayalım.

```python
# Import df for a horizontal bar plot showing the budgets of all 8 Star Wars movies
# 8 Star Wars filminin tümünün bütçelerini gösteren yatay bir çubuk grafiği için df'yi içe aktarın
ax = star_wars_budget_gross_plot.plot(kind='barh', x='movie_title_', y='budget', color='gold')

# Despine
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Set Title
ax.set_title('Star Wars Movies - Budget (in order of release date)')

# Set x-axis label
ax.set_xlabel("Budget (in millions)", labelpad=20, weight='bold', size=12)

# Set y-axis label
ax.set_ylabel("Movie Title", labelpad=20, weight='bold', size=12)

# Set xticks for budget
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 350000000))
ax.set_xticks([0, 50000000, 100000000, 150000000, 200000000, 250000000, 300000000, 350000000])
ax.set_xticklabels(['0', '50M', '100M', '150M', '200M', '250M', '300M', '350M'])


# Import df for a horizontal bar plot showing the gross of all 8 Star Wars movies
# 8 Star Wars filminin tümünün brütünü gösteren yatay bir çubuk grafiği için df'yi içe aktarın
star_wars_gross = star_wars_budget_gross_plot[['movie_title_','gross']]
star_wars_gross_plot = star_wars_gross.sort_index(ascending=False)
ax= star_wars_gross_plot.plot(kind='barh', x='movie_title_', y='gross', color='gold')


# Set Title
ax.set_title('Star Wars Movies - Gross (in order of release date)')

# Set x-axis label
ax.set_xlabel("Gross US & CANADA (in millions)", labelpad=20, weight='bold', size=12)

# Set y-axis label
ax.set_ylabel("Movie Title", labelpad=20, weight='bold', size=12)

# Set xticks for gross
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 1000000000))
ax.set_xticks([0, 100000000, 200000000, 300000000, 400000000, 500000000, 600000000, 700000000, 800000000,
              900000000, 1000000000])
ax.set_xticklabels(['0', '100M', '200M', '300M', '400M', '500M', '600M', '700M', '800M', '900M', '1B'])

plt.show()
```

We can see how the budgets get progressively larger between each of the three trilogies. I am quite surprised to see that the prequels grossed about the same as the original trilogy, despite the fact that they came out some 16–22 years after Episode VI: Return of the Jedi. We would expect that with two generations of Star Wars fans in combination with higher box office ticket prices that the prequels would have grossed much more than the original trilogy.

To get a more accurate comparison of the gross of the prequels vs. the original trilogy, let's calculate the gross mean of all 3 trilogies along with the mean budget. From there we can assess the mean net profit.

Below is a print out of the calculated mean values.

-Üç üçlemenin her biri arasında bütçelerin nasıl giderek büyüdüğünü görebiliriz. Bölüm VI: Jedi'nin Dönüşü'nden 16-22 yıl sonra çıkmış olmalarına rağmen, ön bölümlerin orijinal üçlemeyle hemen hemen aynı hasılat elde ettiğini görmek beni oldukça şaşırttı. İki nesil Star Wars hayranının daha yüksek gişe bilet fiyatlarıyla birleşmesi ile ön oyunların orijinal üçlemeden çok daha fazla hasılat elde etmesini beklerdik.

Orijinal üçlemeyle ön dizilerin brütünün daha doğru bir karşılaştırmasını elde etmek için, ortalama bütçeyle birlikte 3 üçlemenin de brüt ortalamasını hesaplayalım. Oradan ortalama net karı değerlendirebiliriz.

Aşağıda hesaplanan ortalama değerlerin bir çıktısı bulunmaktadır.

```python
import statistics
mean_budget_originals = [11000000.0, 18000000.0, 32500000.0]
mean_budget_prequels = [115000000.0, 115000000.0, 113000000.0]
mean_budget_sequels = [245000000.0, 317000000.0]

org_b = statistics.mean(mean_budget_originals) 
pre_b = statistics.mean(mean_budget_prequels)
seq_b = statistics.mean(mean_budget_sequels) 

print("Mean Budget of Episodes IV, V, & VI :", org_b)
print("Mean Budget of Episodes I, II, & III :", pre_b)
print("Mean Budget of Episodes VII & VIII :", seq_b)

mean_gross_originals = [460935665.0, 290158751.0, 309125409.0]
mean_gross_prequels = [474544677.0, 310675583.0, 380262555.0]
mean_gross_sequels = [936662225.0, 620181382.0]

org_g = statistics.mean(mean_gross_originals) 
pre_g = statistics.mean(mean_gross_prequels)
seq_g = statistics.mean(mean_gross_sequels) 

print("Gross mean of Episodes IV, V, & VI :", org_g)
print("Gross mean of Episodes I, II, & III :", pre_g)
print("Gross mean of Episodes VII & VIII", seq_g)

net_org = org_g - org_b
net_pre = pre_g - pre_b
net_seq = seq_g - seq_b

print("Mean Net Profit of Episodes IV, V, VI :", net_org)
print("Mean Net Profit of Episodes I, II, III :", net_pre)
print("Mean Net Profit of Episodes VII, VIII :", net_seq)
```

The gross mean of the prequels was 388.5 million, only 35.1 million more than the gross mean of the original trilogy of 353.4 million. That's about 10% growth. Nonetheless, the mean net profit of the original trilogy, with their much lower budget, ended up greater than the prequels by about 58 million dollars.

The jump from the prequels to the sequels was substantially more, going from 388.5 million for Episodes I-III to 778.4 million for Episodes VII and VIII. Thats 103% growth, more than double.

Let us now plot the data in a scatter plot to get a deeper look at the relationship between budget and gross for all 8 movies.


-Prequellerin brüt ortalaması 388,5 milyondu, orijinal üçlemenin 353,4 milyonluk brüt ortalamasından sadece 35.1 milyon daha fazla. Bu yaklaşık %10'luk bir büyüme. Bununla birlikte, çok daha düşük bütçeleriyle orijinal üçlemenin ortalama net karı, önceki filmlerden yaklaşık 58 milyon dolar daha fazla oldu.

Bölüm I-III için 388,5 milyondan Bölüm VII ve VIII için 778,4 milyona çıkarak, ön dizilerden devam filmlerine geçiş önemli ölçüde daha fazlaydı. Bu %103 büyüme, iki katından fazla.

Şimdi, 8 filmin tamamı için bütçe ve brüt arasındaki ilişkiye daha derin bir göz atmak için verileri bir dağılım grafiğinde çizelim.

```python
# Import df for a scatter plot displaying budget vs. gross of all 8 movies
# 8 filmin tümünün bütçesine karşı brüt değerini gösteren bir dağılım grafiği için df'yi içe aktarın

ax=star_wars_budget_gross_plot.plot(kind='scatter', x='budget', y='gross', color='#2E80D0')
N=8

# Set labels
ax.set_xlabel('Budget (millions)', labelpad=20, weight='bold', size=12)
ax.set_ylabel('Gross US & CANADA (millions)', labelpad=20, weight='bold', size=12)
ax.set_title('Star Wars Movies I-VIII - Budget vs. Gross')

# Set xticks for budget
ax.set_xticks([0, 50000000, 100000000, 150000000, 200000000, 250000000, 300000000, 350000000])
ax.set_xticklabels(['0', '50M', '100M', '150M', '200M', '250M', '300M', '350M'])

# Set yticks for gross
ax.set_yticks([0, 100000000, 200000000, 300000000, 400000000, 500000000, 600000000, 700000000, 800000000,
              900000000, 1000000000])
ax.set_yticklabels(['0', '100M', '200M', '300M', '400M', '500M', '600M', '700M', '800M', '900M', '1B'])

# Set legend
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

The cluster of three on the left are Episodes IV, V, and VI, the original trilogy. The cluster of three in the middle of the plot is Episodes I, II, and III, the prequels. The two points in the top right corner are Episodes VII and VIII, the sequels.

Like we determined in the summary statistics above, Episodes IV-VI appear to have grossed similarly to the prequels, but given their smaller budgets, had a significantly higher net profit than the prequels. In the upper right corner, we have a very interesting distribution. Episode VII grossed nearly 1 billion dollars in the box office, with a 250 million dollar budget. Episode VII had more than twice the budget as the prequels and rightfully grossed more than twice the mean of the prequels. With three generations of Star Wars lovers who went to see Episode VII in the theater, we would have expected at least this much growth.

Something, however, went wrong with Episode VIII. We would have expected, based on the first two trilogies that Episode VIII would have clustered within a close vicinity of Episode VII. If you look closely, you will see that the budget was around 315 million for Episode VIII, about 65 million more than what was spent on Episode VII. It's quite possible that Disney decided to spend more on Episode VIII based on how well the previous movie did. However, Episode VIII grossed little more than 600 million, significantly less than Episode VII's 936.6 million. We can see the relationships between budget and gross a little differently by a arranging the data in a stacked bar plot. Perhaps the Disney executives could have predicted Episode VIII's less than optimal gross by seeing the pattern set by the other two trilogies.

```python
# Import df for a stacked bar chart displaying discrepancy between budget and gross of all 8 movies
# Tüm 8 filmin bütçesi ile brüt arasındaki tutarsızlığı gösteren bir yığın çubuk grafik için df'yi içe aktarın

ax=star_wars_arragned

# Identify the variables
Episode=['IV','V','VI','I','II', 'III', 'VII', 'VIII']
Budget_Gross=['budget','gross']
pos = np.arange(len(Episode))
y1= ax['budget']
y2= ax['gross']

# Plot a stacked barchart: bottom columns black for budget, top columns gold for gross
# Yığılmış bir çubuk grafik çizin: bütçe için alt sütunlar siyah, brüt için üst sütunlar altın

plt.bar(pos, y1, color='black', label='budget')
plt.bar(pos, y2, color='gold', bottom =y1, label = 'gross')

# Set xticks for episode number and y ticks for budget/gross values
# Bölüm numarası için xticks ve bütçe/brüt değerler için y ticks'i ayarlayın

plt.xticks(pos, Episode)
plt.yticks(np.arange(0, 1400000000, 100000000), ('0', '100M', '200M', '300M', '400M', '500M','600M', '700M', '800M', '900M', '1B', '1.1B', '1.2B', '1.3B'))
# Set xlabels, ylabels, title labels, and legend
plt.xlabel('Star Wars Episode', fontsize=16)
plt.ylabel('Budget & Gross (in millions)', fontsize=14)
plt.title('Budget vs. Gross of Star Wars Movies',fontsize=18)
plt.legend(Budget_Gross,loc=2)
plt.show()
```

We can really see the discrepancy between budget and gross in the stacked bar plot above. Look at Episode IV and how minuscule it's budget is compared to what it grossed.

In the bar plot we can also start to visualize a trend across all three trilogies that may help us to predict the outcome of Episode IX. It looks like the first episode of each trilogy grossed the most, followed by a dip in the second episode of each trilogy and some recovery in the third.

If we look at Episode VII and VIII, we also find part of a similar trend. From the trend we can make a prediction that Episode IX will gross somewhere between 600 million and 1 Billion dollars and have a budget somewhere between 250 and 350 million dollars.

My theory is that the novelty of the first episode of each trilogy generates more ticket sales than the second and third episodes.
